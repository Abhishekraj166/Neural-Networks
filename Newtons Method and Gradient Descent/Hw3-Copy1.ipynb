```python
# Making the imports
import numpy as np
import scipy.io
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import inv
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
i = np.arange(1, 51).reshape((50, 1))
X = i
ui =-1+2*(np.random.rand(1,50).reshape(50,1))
Y = i+ui
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
m = 0
c = 0

L = 0.0001  # The learning Rate
epochs = 3000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    G= np.multiply(X,(Y - Y_pred))
    H = sum(G)
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (c,m)
Y_pred = m*X + c
plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```

```python
ones = np.ones((50, 1))
Xc = np.transpose(np.hstack([ones,X]))
D = np.transpose(Y)
V = np.dot(D,np.dot((np.transpose(Xc)),inv(np.dot(Xc,np.transpose(Xc)))))
inter = np.matmul((Xc),np.transpose(Xc))
inverse = inv(inter)
W = np.matmul(D,(np.matmul((np.transpose(Xc)),inverse)))
print (transpose(W))
```

```python
m = W.item(0,1)
c = W.item(0,0)
Y_pred = m*X + c
plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
#scipy.io.savemat('C:/Users/rajes/Documents/matlab/neural net/arrdata.mat', mdict={'Y': Y})
```
